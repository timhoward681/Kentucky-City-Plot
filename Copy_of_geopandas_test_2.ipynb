```python
import geopandas
import matplotlib.pyplot as plt
```

```python
gdf = geopandas.read_file("/tl_2020_21_county20.shp")

gdf
```

```python
cities = geopandas.read_file("/tl_2020_21_place.shp")

type(cities)
cities
```
